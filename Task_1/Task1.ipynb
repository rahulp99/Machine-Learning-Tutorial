```python
import numpy as np
import math
import matplotlib.pyplot as plt
from numpy import random
```

```python
def gaussian(x, mean, variance):
    const = 1/(((2*math.pi)**0.5)*sigma)
    y = math.e**((-0.5*(sigma**2))*((x - mean)**2))
    return y*const
```

```python
def likelihood(x, mean):
    p = 1
    for i in x:
        p *= gaussian(i, mean, 1)
    return p
```

```python
mean = 5
variance = 1
```

```python
sigma = math.sqrt(variance)
const = 1/(((2*math.pi)**0.5)*sigma)
```

```python
x = random.normal(mean, sigma, 10)
y = [gaussian(i, mean, sigma) for i in x]
```

```python
plt.plot(x, y, 'ro')
plt.title("Guassian Distribution")
plt.savefig("Gaussian distribution")
plt.show()
```

```python
means = [0]*11
likelihoods = [0]*11
for i in range(0, 11):
    likelihoods[i] = likelihood(x, i)
    means[i] = i
```

```python
plt.plot(means, likelihoods, 'r')
plt.xlabel("Mean")
plt.ylabel("Likelihood")
plt.title("Likelihood curve")
plt.savefig("Likelihood curve")
plt.show()
```

```python
log_likelihood = [math.log(i) for i in likelihoods]
```

```python
plt.plot(means, log_likelihood, 'r')
plt.xlabel("Mean")
plt.ylabel("Log Likelihood")
plt.title("Log Likelihood curve")
plt.savefig("Log Likelihood curve")
plt.show()
```
